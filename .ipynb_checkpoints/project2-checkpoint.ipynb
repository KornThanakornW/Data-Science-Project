## คัดกรองข้อมูล
นำคอลัมน์ที่ไม่ได้ใช้หรือไม่จำเป็นออก เพื่อลดขนาดตารางและให้ง่ายต่อการวิเคราะห์

```python
accident = accident.drop(columns=['Unnamed: 0'])
vehicle = vehicle.drop(columns=['Unnamed: 0'])
vision = vision.drop(columns=['Unnamed: 0'])
damage = damage.drop(columns=['Unnamed: 0'])
```

ืเปลี่ยน timestamp_of_crash ให้เป็น type datetime

```python
accident['timestamp_of_crash'] = pd.to_datetime(accident.timestamp_of_crash)
vehicle['timestamp_of_crash'] = pd.to_datetime(vehicle.timestamp_of_crash)
```

แก้ไขข้อมูลหมายเลขวันของสัปดาห์ให้กลายเป็นชื่อวัน เพื่อให้ง่ายต่อการนำไปแสดง

```python
accident.day_of_week
```

```python
day_of_week = pd.DataFrame({'day_of_week': [1,2,3,4,5,6,7], 'day_of_week_name': ['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat']}).set_index('day_of_week')
```

```python
accident = accident.join(day_of_week, on='day_of_week').drop(columns='day_of_week').rename(columns={'day_of_week_name': 'day_of_week'})
accident.day_of_week = pd.Categorical(accident.day_of_week, categories=['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat'], ordered=True)
accident.day_of_week
```

รวมเหตุการณ์ที่เกี่ยวข้องกับสัตว์ให้เป็นเหตุการณ์เดียวกัน

```python
vehicle.loc[vehicle.critical_event_precrash_name.str.contains('Animal')].critical_event_precrash_name.unique()
```

```python
for i in range(len(vehicle.state_number)) :
    if ('Animal' in vehicle['critical_event_precrash_name'][i]) :
        vehicle['critical_event_precrash_name'][i] = 'Relating to Animals'

vehicle.loc[vehicle.critical_event_precrash_name.str.contains('Animal')].critical_event_precrash_name.unique()
```

## วิเคราะห์ข้อมูลเชิงสำรวจเบื้องต้น

### จำนวนการเกิดอุบัติเหตุในแต่ละวันของสัปดาห์

```python
crash_of_week = accident.groupby('day_of_week').count().state_number.to_frame()
crash_of_week = crash_of_week.rename(columns={'state_number': 'number_of_accident'})
crash_of_week
```

```python
graph = sns.relplot(data=crash_of_week, x='day_of_week', y='number_of_accident', kind='line')
```

จากกราฟจะพบว่า อุบัติเหตุเกิดขึ้นมากในช่วงวันเสาร์-อาทิตย์ โดยอุบัติเหตุในวันเสาร์มีจำนวนครั้งการเดิดที่มากที่สุด และเกิดขึ้นน้อยในช่วงกลางๆของสัปดาห์ โดยเฉพาะวันจันทร์และวันอังคาร

### จำนวนอุบัติเหตุที่เกิดขึ้นในแต่ละเดือน

```python
month_name = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
crash_per_month = accident.groupby('month_of_crash').agg({'state_name': pd.Series.count}).rename(columns={'state_name' : 'number_of_accident'})
crash_per_month['month_name'] = month_name
crash_per_month
```

```python
sns.relplot(data=crash_per_month, x='month_name', y='number_of_accident', kind='line')
```

จากกราฟจะเห็นว่าช่วงเดือนกรกฎาคม ถึงเดือนตุลาคม จะมีการเกิดอุบัติเหตุที่ค่อนข้างสูง เมื่อเทียบกับเดือนอื่นๆ

ตารางด้านบนแสดงถึง 10 อันดับ เส้นทางที่มีการเกิดอุบัติเหตุมากที่สุดเรียงจางมากไปน้อย
โดยเส้นทางที่ติดสามอันดับแรก ไดแก่ 
1. I-10 (Interstate 10) เป็นทางหลวงอินเตอร์สเตตที่ตั้งอยู่ทางตอนใต้ของประเทศสหรัฐอเมริกา ทอดตัวจากตะวันออกไปตะวันตก จากชายฝั่งมหาสมุทรแปซิฟิกซึ่งต่อจากทางหลวงแคลิฟอร์เนีย หมายเลข 1 ในเมืองแซนตามอนิกา รัฐแคลิฟอร์เนีย มุ่งสู่อินเตอร์สเตต 95 เมืองแจ็กสันวิลล์ รัฐฟลอริดา 
 <br>( สืบค้นเมื่อ 24/02/2021 https://th.wikipedia.org/wiki/อินเตอร์สเตต_10 )
2. I-95 (Interstate 95) เป็นทางหลวงอินเตอร์สเตตสายหลักในพื้นที่ชายฝั่งตะวันออกของสหรัฐ มีเส้นทางขนานไปกับชายฝั่งมหาสมุทรแอตแลนติก และทางหลวงสหรัฐหมายเลข 1 จากรัฐฟลอริดาไปจนถึงรัฐเมน โดยทั่วไปแล้ว อินเตอร์สเตต 95 จะผ่านเมืองใหญ่ที่อยู่ริมทะเลและพื้นที่ปริมณฑล ได้แก่ ในทางตะวันออกเฉียงใต้ของสหรัฐ เช่น ไมแอมี แจ็กสันวิลล์ ซาวันนาห์ เฟย์เอตต์วิลล์ และริชมอนด์ ส่วนในทางแอตแลนติกกลางไปจนถึงตะวันออกเฉียงเหนือ เช่น วอชิงตัน ดี.ซี. บัลติมอร์ วิลมิงตัน ฟิลาเดลเฟีย นครนิวยอร์ก นิวเฮเวน โพรวิเดนซ์ บอสตัน และพอร์ตแลนด์
 <br>( สืบค้นเมื่อ 24/02/2021 https://th.wikipedia.org/wiki/อินเตอร์สเตต_95 )
3. I-75 (Interstate 75) เป็นทางหลวงอินเตอร์สเตตสายหลักในภูมิภาคเกรตเลกส์และภูมิภาคสหรัฐตะวันออกเฉียงใต้ของสหรัฐอเมริกา หมายเลขที่ลงท้ายด้วยเลข 5 หมายความว่าเป็นทางหลวงสายหลักของประเทศในแนวเหนือ–ใต้ มีเส้นทางจากตอนใต้ของรัฐฟลอริดาไปจนถึงทางเหนือของรัฐมิชิแกน
 <br>( สืบค้นเมื่อ 24/02/2021 https://th.wikipedia.org/wiki/อินเตอร์สเตต_75 )

## ความน่าสนใจที่ได้จากชุดข้อมูล

### จำนวนการเกิดอุบัติเหตุแยกตามช่วงเวลาของทุกวัน

```python
time_crash = accident.groupby(pd.Grouper(key="timestamp_of_crash", freq="1H")).count().sort_values(by="state_number", ascending=False)
```

```python
sort_by_hr = pd.DataFrame([[0] for i in range(24)], columns=["count"]).rename_axis("Hour")
for row, col in time_crash.iterrows():
    sort_by_hr['count'][row.hour] += col['state_number']
sort_by_hr = sort_by_hr.sort_values('count', ascending=False)
```

```python
sns.relplot(data=sort_by_hr, kind='line')
```

พบว่าช่วงเวลาที่มักจะเกิดอุบัติเหตุมากที่สุดคือช่วงเวลาหลังจากเลิกงานเป็นต้นไป 16:00 - 21:00 น.  
และช่วงเวลาที่เกิดอุบัติเหตุน้อยที่สุดคือช่วงเช้า-มืด 00:00 - 04:00 น.

### วิเคราะห์ความสัมพันธ์ของระดับความเสียหายที่เกิดขึ้นกับเหตุการณ์ก่อนเกิดอุบัติเหตุ

```python
reason_damage = vehicle.join(damage.drop(columns=['state_number', 'vehicle_number']).set_index('consecutive_number'), on='consecutive_number').loc[:,['consecutive_number','damaged_areas','critical_event_precrash_name']]
reason_damage
```

ค่าเฉลี่ยความเสียหายที่เกิดขึ้นกับยานพาหนะเทียบกับจำนวนของอุบัติเหตุที่เกิดขึ้นแยกตามเหตุการณ์นั้น ๆ

```python
reason_damage_mean = reason_damage.groupby(by='critical_event_precrash_name').agg({'consecutive_number': pd.Series.count, 'damaged_areas': pd.Series.mean}).sort_values('consecutive_number', ascending=False)
reason_damage_mean
```

จากตารางจะเห็นได้ว่าค่าเฉลี่ยของเหตุการณ์ที่่เกิดอุบัติเหตุ โดยส่วนใหญ่แล้วยานพาหนะไม่ได้รับความเสียมากนัก

```python
sns.relplot(data=reason_damage_mean.head(10), x='consecutive_number', y='damaged_areas', hue='critical_event_precrash_name')
```

### วิเคราะห์อุบัติเหตุที่เกิดจากการดื่มสุรา

```python
drunk_accident = accident.loc[accident.number_of_drunk_drivers > 0, ['state_name', 'number_of_drunk_drivers']]
drunk_accident = drunk_accident.groupby('state_name').count().rename(columns={'number_of_drunk_drivers': 'number_of_drunk_cases'})
drunk_accident.sort_values('number_of_drunk_cases', ascending=False).head(5)
```

จากตารางพบว่ารัฐที่มีอุบัติเหตุทางรถยนต์ที่เกิดจากการดื่มสุรามากที่สุด 5 ลำดับ เรียงตามจำนวนการเกิดมากไปน้อย ได้แก่ 'California', 'Texas', 'Florida', 'North Carolina', 'Pennsylvania'

```python
drunk_accident.sort_values('number_of_drunk_cases', ascending=True).head(5)
```

จากตารางพบว่ารัฐที่มีอุบัติเหตุทางรถยนต์ที่เกิดจากการดื่มสุราน้อยที่สุด 5 ลำดับ เรียงตามจำนวนการเกิดน้อยไปมาก ได้แก่ 'District of Columbia', 'Vermont', 'Rhode Island', 'Alaska', 'Hawaii'

```python
g = sns.catplot(data=drunk_accident.reset_index(), x='state_name', y='number_of_drunk_cases', kind='bar', aspect=2.5)
g.set_xticklabels(rotation=75)
```

จากตารางและกราฟด้านบนแสดงให้เห็นว่า รัฐที่มีอุบัติเหตุทางรถยนต์จากการดื่มสุรามากที่สุด คือ รัฐแคลิฟอร์เนีย และมีจำนวนรองลงมาคือ รัฐเท็กซัส และรัฐฟลอริา ตามลำดับ

```python
drunk_number = len(accident.loc[accident.number_of_drunk_drivers > 0])
non_drunk_number = len(accident.loc[accident.number_of_drunk_drivers == 0])
```

```python
labels = ['Drunk', 'Normal']
sizes = [drunk_number, non_drunk_number]
explode = (0.1, 0.1)  
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  
plt.tight_layout()
plt.show()
```
